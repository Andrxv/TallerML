Esta tabla tiene como tipo de dato **DataFrame**, mientras que las columnas tienen como tipo de dato **Series**. Ya despues de tener el DataFrame cargado podemos hacer un sin númeroo de cosas como:

```python
HR_df.describe()
```

- Obtener datos de una columna especifica.

```python
#Imprime los valores unicos de la columna.
HR_df.dept.unique()
```

- Obtener filas que cumplen con condiciones especificas.

```python
left_company = HR_df.loc[HR_df.left == 1]
left_company.describe()
```

- Modificar datos

```python
percentage_HR = HR_df.copy() #Crear una copia del dataframe, para no modificar la original.
percentage_HR[["satisfaction_level","last_evaluation"]] = (percentage_HR[["satisfaction_level","last_evaluation"]] 
                                                           * 100).astype(str) + "%"
percentage_HR.head()
```

- Agrupar datos

Para la creación de gráficos y visualizaciones en Python, se utilizan dos librerias comúnmente:

- Matplotlib: Creación de visualizaciones a bajo nivel.
- Seaborn: Abstrae y da estilo a las visualizaciones. Esta hecha sobre Matplotlib.

Para una compración entre ambas:

```python
import matplotlib.pyplot as plt
#Scatterplot
HR_viz= HR_df.sample(frac = 0.01) #fracción más pequeña del dataset.
plt.scatter(HR_viz.satisfaction_level,HR_viz.average_montly_hours)
plt.xlabel("Nivel de satisfacción")
plt.ylabel("Horas promedio mensuales")
plt.title("Nivel de satisfacción en funcion de horas trabajadas")
plt.show()
```

El equivalente en seaborn:

```python
import seaborn as sns
plot = sns.regplot(x = "satisfaction_level", y = "average_montly_hours", data = HR_viz)
plt.title("Nivel de satisfacción en funcion de horas trabajadas")
plt.show()
```

Como se puede apreciar, se pueden combinar Matplotlib con Seaborn para obtener la complejidad deseada en las visualizaciones. Tambien, se puede modificar aspectos visuales de la gráfica con **RcParams**.

```python
from matplotlib import rcParams 
#Tamaño de la figura.
rcParams['figure.figsize'] = (8, 6)
#Resolucion de la figura.
rcParams['figure.dpi'] = 100
#Tamaño de letra.
rcParams['font.size'] = 14
histograma = sns.distplot(HR_df.average_montly_hours,kde = False,bins = 5)
plt.show()
```

Y además de los tipos de gráficos vistos anteriormente tambien están:

- Bar Chart

```python
bar = sns.barplot(x = "salary", y = "satisfaction_level",hue = "left",data = HR_df)
plt.xlabel("Salario")
plt.ylabel("Nivel de satisfacción promedio")
plt.xticks(range(len(HR_df.salary.unique())),["Bajo","Medio","Alto"]) # Reemplazar el texto del eje X.
plt.legend(title = "Se fue")
plt.show()
```

- Box Plots

```python
box = sns.boxplot(x = "time_spend_company", y = "average_montly_hours", data=  HR_df)
plt.xlabel("Años en la Compañia")
plt.ylabel("Horas mensuales")
plt.show()
```

- Mapas de correlación o heatmaps.

```python
corr = HR_df.corr() #Crear matriz de correlación
rcParams["font.size"] = 10
rcParams["figure.dpi"] = 125
plt.matshow(corr)
ax = sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values,
            annot = True)
plt.xticks(rotation = 70) # rotar los nombres del eje X.
plt.show()
```

Esta libreria realiza todo tipo de operaciones con vectores y matrices, de manera rápida y eficiente.

```python
import numpy as np
vector1 = np.array([1,2,3])
vector2 = np.array([1,-1,1])
print(vector1 + vector2)

print(np.dot(vector1,vector2))#Producto punto
```

Tambien permite crear distribuciones estadisticas o de numeros aleatorios.

```python
random = np.random.random(10) # Distribucion aleatoria de 10 numeros.
print(random)
binomial = np.random.binomial(1,0.5, size =10)#Distribucion binomial de tamaño 10.
print(binomial)
```

```python
normal = np.random.normal(size = 10) #Distribución normal aleatoria
normal
```

Y funciones matematicas y trigonómetricas.

```python
#Grafica del Seno
numeros=  np.arange(-5,5,0.1) #Intervalo de numeros de -5 a 5, a distancia 0.1 cada uno.
y = np.sin(numeros)
plt.plot(numeros,y)
plt.show()
```
